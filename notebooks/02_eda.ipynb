# Eksploracyjna analiza danych (EDA)

W niniejszym notatniku przedstawiono eksploracyjną analizę zbioru danych dotyczących sprzedaży w handlu elektronicznym.

Celem tej analizy jest:
- zrozumienie struktury zbioru danych,
- analiza kluczowych wskaźników sprzedaży,
- identyfikacja trendów sprzedaży,
- zbadanie zachowań produktów i klientów.

## Import bibliotek

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Załaduj oczyszczony zbiór danych

Załadowanie zbioru danych przygotowanych podczas etapu czyszczenia danych.

```python
df = pd.read_csv("../data/processed/sales_cleaned.csv")

df.head()
```

## Przegląd zbioru danych

Sprawdzenie rozmiaru zbioru danych, struktury i typu danych.

```python
print("Dataset shape:", df.shape)

df.info()
```

## Podstawowe statystyki opisowe

```python
df.describe()
```

## KPI sprzedażowe
Obliczenie podstawowych wskaźników sprzedaży

```python
total_revenue = df["revenue"].sum()
total_orders = df["order_id"].nunique()
total_customers = df["customer_unique_id"].nunique()

avg_order_value = (
    df.groupby("order_id")["revenue"]
    .sum()
    .mean()
)

print("Total revenue:", total_revenue)
print("Total orders:", total_orders)
print("Total customers:", total_customers)
print("Average order value:", avg_order_value)
```

### Interpretacja

Wskaźniki te zapewniają ogólny przegląd wyników działalności.

Kluczowe wskaźniki:
- całkowity przychód wygenerowany przez sklep
- liczba unikalnych zamówień
- liczba klientów
- średnia wartość zamówienia

## Sprzedaż w czasie
Przeanalizowanie jak zmieniają sie przychody w czasie

```python
monthly_sales = (
    df.groupby("order_month")["revenue"]
    .sum()
    .reset_index()
)

monthly_sales.head()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    data=monthly_sales,
    x="order_month",
    y="revenue"
)

plt.xticks(rotation=45)

plt.title("Revenue over time")

plt.tight_layout()

plt.savefig("../reports/figures/monthly_revenue.png")

plt.show()
```

### Interpretacja

Wykres pokazuje zmiany przychodów w czasie.

Pozwala to zidentyfikować:
- wzorce sezonowe
- trendy wzrostowe
- potencjalne spadki sprzedaży

## Top produktów

```python
top_products = (
    df.groupby("product_id")
    .agg(
        revenue=("revenue","sum"),
        quantity=("product_id","count")
    )
    .sort_values("revenue",ascending=False)
    .head(10)
    .reset_index()
)

top_products
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=top_products,
    x="revenue",
    y="product_id"
)

plt.title("Top 10 products by revenue")

plt.tight_layout()

plt.savefig("../reports/figures/top_products.png")

plt.show()
```

## Analiza klientów
Zidentyfikowanie klientów generujących największe przychody.

```python
top_customers = (
    df.groupby("customer_unique_id")
    .agg(
        orders=("order_id","nunique"),
        revenue=("revenue","sum")
    )
    .sort_values("revenue",ascending=False)
    .head(10)
    .reset_index()
)

top_customers
```

## Podsumowanie

Analiza eksploracyjna ujawniła kilka ważnych wniosków:

- przychody zmieniają się w czasie
- niektóre produkty generują znacznie wyższą sprzedaż
- niewielka grupa klientów ma nieproporcjonalnie duży udział w przychodach

Obserwacje te zostaną dokładniej zbadane w kolejnych analizach.

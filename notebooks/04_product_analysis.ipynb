# Analiza produktów

Celem tego notebooka jest analiza sprzedaży produktów w sklepie internetowym.

Analiza obejmuje:

- identyfikację produktów generujących najwyższy przychód
- identyfikację najczęściej kupowanych produktów
- analizę kategorii produktów
- analizę rozkładu cen produktów

Wyniki tej analizy pomagają zrozumieć strukturę sprzedaży i wskazać najważniejsze elementy oferty sklepu.

## Import bibliotek

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Wczytanie danych

```python
df = pd.read_csv("../data/processed/sales_cleaned.csv")

df.head()
```

## Produkty generujące najwyższy przychód

```python
top_products_revenue = (
    df.groupby("product_id")
    .agg(
        revenue=("revenue","sum"),
        quantity=("product_id","count")
    )
    .sort_values("revenue",ascending=False)
    .head(10)
    .reset_index()
)

top_products_revenue
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=top_products_revenue,
    x="revenue",
    y="product_id"
)

plt.title("Top 10 produktów według przychodu")

plt.tight_layout()

plt.savefig("../reports/figures/top_products_revenue.png")

plt.show()
```

### Interpretacja

Analiza pokazuje produkty generujące najwyższy przychód.

Produkty te mają największy wpływ na całkowity przychód sklepu i mogą być kluczowe z perspektywy strategii sprzedaży.

```python
top_products_quantity = (
    df.groupby("product_id")
    .agg(
        quantity=("product_id","count"),
        revenue=("revenue","sum")
    )
    .sort_values("quantity",ascending=False)
    .head(10)
    .reset_index()
)

top_products_quantity
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=top_products_quantity,
    x="quantity",
    y="product_id"
)

plt.title("Top 10 najczęściej kupowanych produktów")

plt.tight_layout()

plt.savefig("../reports/figures/top_products_quantity.png")

plt.show()
```

### Interpretacja

Najczęściej kupowane produkty nie zawsze generują największy przychód.

Porównanie liczby sprzedaży i przychodu pozwala zidentyfikować produkty o wysokiej wartości lub wysokim wolumenie sprzedaży.

## Analiza kategorii produktów

```python
category_sales = (
    df.groupby("product_category_name")
    .agg(
        revenue=("revenue","sum"),
        quantity=("product_id","count")
    )
    .sort_values("revenue",ascending=False)
    .head(10)
    .reset_index()
)

category_sales
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=category_sales,
    x="revenue",
    y="product_category_name"
)

plt.title("Top 10 kategorii produktów według przychodu")

plt.tight_layout()

plt.savefig("../reports/figures/top_categories_revenue.png")

plt.show()
```

### Interpretacja

Analiza kategorii produktów pozwala określić segmenty oferty, które generują największą część przychodu.

Informacja ta może być wykorzystana przy podejmowaniu decyzji dotyczących rozwoju oferty sklepu.

## Rozkład cen produktów

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df["price"],
    bins=50
)

plt.title("Rozkład cen produktów")

plt.tight_layout()

plt.savefig("../reports/figures/product_price_distribution.png")

plt.show()
```

### Interpretacja

Rozkład cen produktów pozwala zrozumieć strukturę cen w ofercie sklepu.

Można sprawdzić:

- czy większość produktów znajduje się w określonym przedziale cenowym
- czy występują produkty o bardzo wysokiej cenie

## Podsumowanie

Analiza produktów pozwoliła zidentyfikować:

- produkty generujące największy przychód
- produkty sprzedające się w największej liczbie
- najważniejsze kategorie produktów
- strukturę cen produktów
